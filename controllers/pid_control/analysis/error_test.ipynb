```python
import glob

import matplotlib.pyplot as plt
import pandas as pd
```

## Proportional Control

```python
p_dfs = {}
for f in glob.glob('../out_data/pid_kp_*_ki_0.000_kd_0.000.csv'):
    file_name = f.split('\\')[1]
    print(file_name)
    p_dfs[file_name] = pd.read_csv(f)
```

```python
p_dfs['pid_kp_0.015_ki_0.000_kd_0.000.csv'].head()
```

```python
fig, ax = plt.subplots(figsize=(14, 5))

setpoint = next(iter(p_dfs.values()))["setpoint"].iloc[0]

for name, df in p_dfs.items():
    ax.plot(df["time"], df["measured"], label=name)

ax.axhline(y=setpoint, color='r', linestyle='--', label='setpoint')
ax.set_xlabel("time [s]")
ax.set_ylabel("error")
ax.legend()
plt.tight_layout()
plt.show()
```
